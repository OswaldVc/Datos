# Seaborn

Seaborn es otra librera de visualizacin para python. Posee grficos similares a matplotlib, pero a ms alto nivel. Adems, podemos integrar seaborn con matplotlib.


```python
from locale import normalize
from datetime import datetime
from datetime import date
import matplotlib.pyplot as plt
import missingno as msno
## Data cleaning y data wrangling
import pandas            as pd
import numpy             as np
## Visualizacin
import matplotlib.pyplot as plt
import seaborn           as sns
import plotly.express    as px
import warnings

sns.set_style("darkgrid") ## Estilo para los grficos
warnings.filterwarnings("ignore")
plt.style.use("ggplot")
```

```python
data = sns.load_dataset('mpg')
data.head()
```

```python

sns.set_theme(style="darkgrid")
sns.relplot(x = 'weight', y = 'horsepower', data = data, height=8)
```

Si bien los puntos se representan en dos dimensiones, se puede agregar otra dimensin al grfico coloreando los puntos segn una tercera variable. En seaborn, esto se implementa con el parmetro 'hue'.

```python
sns.relplot(x = 'weight', y = 'horsepower', data = data, height=8, hue = 'origin')
```

An podemos aadir una cuarta dimensin con el parmetro style

```python
sns.relplot(x = 'weight', y = 'horsepower', data = data, height=8, hue = 'origin', style = 'cylinders')
```

Podemos aplicar la misma lgica, pero usando el tamao del punto en lugar de la forma

```python
sns.relplot(x = 'weight', y = 'horsepower', data = data, height=8, hue = 'origin', size = 'cylinders', sizes = (15, 200))
```

Tambin podemos usar el grfico de lineas en lugar de puntos, bien sea para enfatizar una cada o porque nos guste ms ese estilo.

```python
sns.relplot(x = 'weight', y = 'horsepower', data = data, height=7, kind = 'line')
```

```python
hp_mean = data.groupby('cylinders')['horsepower'].mean()
sns.relplot(x = 'cylinders', y = 'horsepower', data = data, height=7, kind = 'line', hue = 'origin', ci = 'sd')
```

Tambin podemos trabajar con fechas de la misma forma que lo hemos hecho hasta ahora. Veamos un ejemplo

```python
df = pd.DataFrame(dict(time=pd.date_range("2020-01-01", periods=200),
                       value=np.random.randn(200).cumsum()))
g = sns.relplot(x="time", y="value", kind="line", data=df, height = 8)
g
```
